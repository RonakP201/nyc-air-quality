```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_air_quality.csv')
df['Year'] = pd.to_datetime(df['Start_Date']).dt.year
```

```python
# Plot 1: Average pollutant level by year
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Year', y='Value', hue='Pollutant')
plt.title('Pollutant Levels by Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Distribution by location
plt.figure(figsize=(12, 6))
top_locations = df['Location'].value_counts().nlargest(10).index
sns.boxplot(data=df[df['Location'].isin(top_locations)], x='Location', y='Value')
plt.xticks(rotation=90)
plt.title('Pollutant Level Distribution by Top Locations')
plt.tight_layout()
plt.show()
```
